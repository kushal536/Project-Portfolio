```python

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
from sklearn.linear_model import LinearRegression
%matplotlib inline

bk=pd.read_csv('Book2.csv')
#print(bk.head())

TRAIN_INPUT1 = list()

TRAIN_OUTPUT1 = list()

TRAIN_INPUT1=list(np.array(bk['Age']).reshape(-1,1))

#print(TRAIN_INPUT1)


TRAIN_OUTPUT1=list(np.array(bk['Salary']).reshape(-1,1))

#print(TRAIN_OUTPUT1)


df=pd.DataFrame(zip(TRAIN_INPUT1,TRAIN_OUTPUT1),columns=["Age","Salary"])
#a= print(df)

X = np.array(TRAIN_INPUT1)
#print(X)
y= np.array(TRAIN_OUTPUT1)
#print(y)


predictor = LinearRegression()
predictor.fit(X=TRAIN_INPUT1, y=TRAIN_OUTPUT1)

matrix = np.array(list(map(float, input("Enter your age: ").split())) ).reshape(1, 1) 
X_TEST= matrix
#print(type(X_TEST))
#print(X_TEST)
output = predictor.predict(X=X_TEST)
coefficients = predictor.coef_

df2 = pd.DataFrame(zip(X_TEST,output),columns=["Age","Salary"])
df3=df.append(df2,ignore_index=True)
#print(df3)
df4=df3[["Age","Salary"]].astype("int")
df4.dtypes
print(df4)
df4.to_csv('Book2.csv')

df5=df4.set_index("Age")
ax=df5.plot(kind='bar',figsize=(20, 6), color='#d9534f')
plt.title('Salary by Age')
plt.xlabel('Age')
plt.ylabel('Salary')


plt.rc('xtick', labelsize=14)    # fontsize of the tick labels
plt.rc('ytick', labelsize=14)    # fontsize of the tick labels
plt.rc('legend', fontsize=14) 

for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy() 
    ax.annotate('{:}k'.format(height/1000), (x, y + height + 0.9))
    
plt.show()

df4.describe()
```
